# 通过颜色变化生成新的图片

<span style="color:red">导入必要的库<span>

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def genColorImg(img, delta, thresh=230):
    th = img < thresh
    a = np.zeros(img.shape, dtype='uint8')
    a[th] = delta
    return img + a
```

```python
for i in range(-50, 50):
    newimg = genColorImg(img, i)
    plt.imshow(newimg)
    plt.show()
```

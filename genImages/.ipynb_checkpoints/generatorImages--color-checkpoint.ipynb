```python
def genColorImg(img, delta, thresh=230):
    th = img < thresh
    a = np.zeros(img.shape, dtype='uint8')
    a[th] = delta
    return img + a
```

```python
newimg = genColorImg(img, 20)
plt.imshow(newimg)
plt.show()
```
